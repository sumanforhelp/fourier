### Here we will use arange function and mostly for loop as well as if-else statement to get fourier series. At first we will generate square wave and then continue to get triangular wave.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

sin(n$\theta$)plot for various values of n

```python
fs=20
amp=1
theta=np.arange(0,2*np.pi,1/fs)
```

### lets plot sine wave  with varying periodicity

```python
plt.figure(figsize=[10,4])
for i in range(1,11):
    y=amp/float(i)*(np.sin(i*theta))
    plt.plot(theta,y)
```

### lets plot cosine wave with varying periodicity

```python
plt.figure(figsize=[10,4])
for i in range(1,11):
    ys=amp/float(i)*(np.cos(i*theta))
    plt.plot(theta,ys)
```

# periodicity of blue wave is 2pi
# periodicity of orange line is 2pi/2
# periodicity of red line is 2pi/3
# periodicity of green line is 2pi/4


```python
theta=np.arange(0,20*np.pi,0.01)
plt.figure(figsize=[20,4])
for i in range(1,10):
    y=float(amp)/float(i)*(np.sin(theta/i))
    plt.plot(theta,y)
```

# fourier transform


![img](https://mathworld.wolfram.com/images/equations/FourierSeriesSquareWave/NumberedEquation3.gif)

```python
plt.figure(figsize=[15,4])
L=1
ns=100
x=np.arange(0,2,1/ns)
s=[ p for p in range(14)]
b=1
for n in s:
    if n%2!=0:
        y=(4/np.pi)*(1/n)*np.sin(n*np.pi*x/L)
        plt.plot(x,y)
        if n==1:
            y1=(4/np.pi)*(1/n)*np.sin(n*np.pi*x/L)
            plt.plot(x,y1)
            yy=y1
        else:
            y1=yy+(4/np.pi)*(1/n)*np.sin(n*np.pi*x/L)
            plt.plot(x,y1)
            yy=y1
        b=b+1
```
